```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft2, ifft2, fftshift, ifftshift
from PIL import Image

# Load the uploaded image
image_path = 'evening.jpeg'
original_image = Image.open(image_path).convert('L')  # Convert to grayscale

# Convert image to numpy array
image_array = np.array(original_image)

# Step 2: Apply Fourier Transform
ft_image = fft2(image_array)              # Perform Fourier Transform
ft_image_shifted = fftshift(ft_image)     # Shift zero frequency component to the center
magnitude_spectrum = np.log(np.abs(ft_image_shifted) + 1)  # Compute magnitude spectrum

# Step 3: Apply Inverse Fourier Transform
ift_image_shifted = ifftshift(ft_image_shifted)  # Shift back to original
reconstructed_image = np.abs(ifft2(ift_image_shifted))  # Inverse Fourier Transform

# Plot Results
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image_array, cmap='gray')
plt.axis('off')

# Magnitude Spectrum
plt.subplot(1, 3, 2)
plt.title("Magnitude Spectrum")
plt.imshow(magnitude_spectrum, cmap='gray')
plt.axis('off')

# Reconstructed Image
plt.subplot(1, 3, 3)
plt.title("Reconstructed Image")
plt.imshow(reconstructed_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
